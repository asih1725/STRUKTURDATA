```python
class PageNode:
  def __init__(self, url):
    self.url = url
    self.next = None
    self.prev = None

class BrowserHistory:
  def __init__(self):
    self.current = None

  def visit_page(self, url):
    new_page = PageNode(url)
    if not self.current:
      self.current = new_page
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page

  def show_history(self):
    if not self.current:
      print("\n Riwayat kosong.")
      return

    print("\n Riwayat Peramban Web:")
    temp = self.current
    while temp.prev:
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.url}")
      temp= temp.next
      count += 1

  def go_back(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f"\n Anda Kembali ke: {self.current.url}")
    else:
      print("\n Tidak bisa kembali, ini adalah halaman pertama.")

  def go_forward(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f"\n Anda maju ke: {self.current.url}")
    else:
      print("\n Tidak bisa maju, ini adalah halaman pertama.")

#MAIN

history= BrowserHistory()

num_pages= int(input("Masukkan jumlah halaman yang akan dikunjungi: "))

for i in range(1, num_pages + 1):
  url = input(f"Masukkan URL halaman ke-{i}: ")
  history.visit_page(url)

history.show_history()


while True:
  command = input("\n Tekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ")

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat peramban web.")
    break
  else:
    print("\n Perintah tidak dikenal")
```

```python
class FolderNode:
    def __init__(self, folder_name):
        self.folder_name = folder_name
        self.next = None
        self.prev = None

class FolderHistory:
    def __init__(self):
        self.current = None

    def visit_folder(self, folder_name):
        new_folder = FolderNode(folder_name)
        if not self.current:
            self.current = new_folder
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder

    def show_history(self):
        if not self.current:
            print("\nRiwayat folder kosong.")
            return

        print("\nRiwayat Navigasi Folder:")
        temp = self.current
        while temp.prev:
            temp = temp.prev
        count = 1

        while temp:
            print(f"{count}. {temp.folder_name}")
            temp = temp.next
            count += 1

    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\n< Kembali ke folder: {self.current.folder_name}")
        else:
            print("\n Tidak bisa kembali, ini adalah folder pertama.")

    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\n> Maju ke folder: {self.current.folder_name}")
        else:
            print("\n Tidak bisa maju, ini adalah folder terakhir yang dikunjungi.")

#MAIN
history = FolderHistory()

num_folders = int(input("Masukkan jumlah folder yang akan dikunjungi: "))

for i in range(1, num_folders + 1):
    folder_name = input(f"Masukkan nama folder ke-{i}: ")
    history.visit_folder(folder_name)

history.show_history()

while True:
    command = input("\nTekan 'b' untuk kembali, 'f' untuk maju, 'x' untuk keluar.\nMasukkan perintah: ")

    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\nKeluar dari riwayat navigasi folder.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```

```python
class SongNade:
  def __init__(self, title):
    self.title = title
    self.next = None

class Playlist:
  def __init__(self):
    self.current = None

  def add_song(self, title):
    new_song = SongNade(title)
    if not self.current:
      self.current = new_song
      self.current.next = self.current
    else:
      temp = self.current
      while temp.next != self.current:
        temp = temp.next
      temp.next = new_song
      new_song.next = self.current

  def show_playlist(self):
    if not self.current:
      print("\n Playlist kosong!")
      return

    print("\n Playlist Lagu Anda:")
    temp = self.current
    count = 1
    while True:
      print(f"{count}. {temp.title}")
      temp = temp.next
      count += 1
      if temp == self.current:
        break

  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f"\n Memutar: {self.current.title}")

#MAIN

playlist = Playlist()

num_songs = int(input("Masukkan jumlah lagu dalam playlist: "))

for i in range(1, num_songs + 1):
  title = input(f"Masukkan judul lagu ke-{i}: ")
  playlist.add_song(title)

playlist.show_playlist()

while True:
  command = input("\n Tekan 'n untuk lagu berikutnya. Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
  if command == 'n':
    playlist.next_song()
  elif command =='x':
    print("\n Keluar dari pemutar musik.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None

  def add_at_beginning(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node

  def add_at_end(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node

  def add_at_position(self, position, data):
    if position < 0:
      self.add_at_beginning(data)
      return
    new_node = Node(data)
    if position == 0:
      new_node.next = self.head
      self.head = new_node
      return
    temp = self.head
    for i in range(position - 1):
      if not temp:
        print("Posisi melebihi panjang daftar!")
        return
      temp = temp.next
    if not temp:
        print("Posisi melebihi panjang daftar!")
        return
    new_node.next = temp.next
    temp.next = new_node

  def delete_at_beginning(self):
    if not self.head:
      print("Daftar kosong, tidak ada yang bisa dihapus!")
      return
    self.head = self.head.next

  def delete_at_end(self):
    if not self.head:
      print("Daftar kosong, tidak ada yang bisa dihapus!")
      return
    if not self.head.next:
      self.head = None
      return
    temp = self.head
    while temp.next.next:
      temp = temp.next
    temp.next = None

  def delete_at_position(self, position):
    if not self.head:
      print("Daftar kosong, tidak ada yang bisa dihapus!")
      return
    if position == 0:
      self.head = self.head.next
      return
    temp = self.head
    for i in range(position - 1):
      if not temp or not temp.next:
        print("Posisi tidak ditemukan!")
        return
      temp = temp.next
    if not temp or not temp.next:
        print("Posisi tidak ditemukan!")
        return
    temp.next = temp.next.next

  def search (self, key):
    temp = self.head
    while temp:
      if temp.data == key:
        print(f"Elemen {key} ditemukan dalam daftar!")
        return
      temp = temp.next
    print(f"Elemen {key} tidak ditemukan dalam daftar.")

  def reverse(self):
    prev = None
    current = self.head
    while current:
      next_node = current.next
      current.next = prev
      prev = current
      current = next_node
    self.head = prev

  def display(self):
    if not self.head:
      print("Daftar kosong!")
      return
    temp = self.head
    while temp:
      print(temp.data, end=" -> ")
      temp = temp.next
    print("None")

#MAIN

II = LinkedList()

num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))

for i in range(num_elements):
  data = int(input(f"Masukkan data ke-{i + 1}: "))
  II.add_at_end(data)

while True:
  print("\n Pilih Operasi:")
  print("1. Tambah elemen di awal")
  print("2. Tambah elemen di akhir")
  print("3. Tambah elemen di posisi tertentu")
  print("4. Hapus elemen di awal")
  print("5. Hapus elemen di akhir")
  print("6. Hapus elemen di posisi tertentu")
  print("7. Cari elemen dalam daftar")
  print("8. Balikkan daftar")
  print("9. Tampilkan daftar")
  print("0. Keluar")

  choice = int(input("\nMasukkan pilihan: "))

  if choice == 1:
    data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
    II.add_at_beginning(data)
  elif choice == 2:
    data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
    II.add_at_end(data)
  elif choice == 3:
    position = int(input("Masukkan posisi untuk menambahkan data: "))
    data = int(input("Masukkan data yang ingin ditambahkan: "))
    II.add_at_position(position, data)
  elif choice == 4:
    II.delete_at_beginning()
  elif choice == 5:
    II.delete_at_end()
  elif choice == 6:
    position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
    II.delete_at_position(position)
  elif choice == 7:
    key = int(input("Masukkan elemen yang ingin dicari: "))
    II.search(key)
  elif choice == 8:
    II.reverse()
    print("Daftar berhasil dibalik!")
  elif choice == 9:
    II.display()
  elif choice == 0:
    print("\n Keluar dari program.")
    break
  else:
    print("\n Pilihan tidak valid, silakan coba lagi!")
```
